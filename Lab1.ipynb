```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
def dydt(x,t):
  k=0.42
  return k*x
t = np.linspace(0,10,100)
solve = odeint(dydt, 1, t=t)
solve = solve.flatten()
%matplotlib inline
plt.plot(t,solve, label = r"$x' =kx$")
plt.grid(True)
plt.legend()
plt.show()
```
